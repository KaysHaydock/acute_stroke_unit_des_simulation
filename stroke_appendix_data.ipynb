## **1. Patient Admission Sources**  
Patients enter the hospital through **two primary pathways**:  
- **New Admissions**: Direct hospital entry for Stroke, TIA, Complex Neurological, and Other Medical Cases.  
- **Transfers from Other Hospitals**: Patients arriving from other hospitals or care facilities.

| Patient Type                     | Admissions (n) | Percentage (%) |
|----------------------------------|---------------|--------------|
| Stroke                           | 1,320         | 54%          |
| Transient Ischemic Attack (TIA)  | 158           | 6%           |
| Complex Neurological             | 456           | 19%          |
| Other Medical Cases              | 510           | 21%          |

---

## **2. Acute Stroke Unit (ASU) Stay**  
Upon admission, patients first enter the **Acute Stroke Unit (ASU)** where they receive initial treatment.

- **Bed Availability**: 10 acute stroke beds.  
- **Mean Length of Stay (LoS)**:  
  - Stroke (No ESD): **7.4 days**  
  - Stroke (ESD): **4.6 days**  
  - Complex Neurological: **4.0 days**  
  - TIA: **1.8 days**  
  - Other: **3.8 days**  

After treatment, patients either **discharge home, transfer to rehab, or enter ESD (Early Supported Discharge).**

---

## **3. Patient Pathways from Acute Stroke Unit (ASU)**  
After the acute phase, patients follow different paths based on their condition.

### **Path 1: Early Supported Discharge (ESD)**
- Patients with mild to moderate strokes may **skip rehab** and receive care at home.  
- **Discharged from ASU:** **300 patients (63% of ESD cases)**  
- **Discharged from Rehab:** **163 patients (37% of ESD cases)**  

### **Path 2: Transfer to Rehabilitation Unit**
- **Patients requiring further recovery** move to the **Rehabilitation Unit**.  
- **Bed Availability**: 12 rehab beds.  
- **Mean LoS in Rehab**:  
  - Stroke (No ESD): **28.4 days**  
  - Stroke (ESD): **30.3 days**  
  - Complex Neurological: **27.6 days**  
  - Other: **16.1 days**  
  - TIA: **18.7 days**  

### **Path 3: Direct Discharge from Acute Unit**
- Patients who **do not need rehab** are discharged home, to care homes, or may experience mortality.

---

## **4. Patient Pathways from Rehabilitation Unit**  
After rehab, patients have **two possible outcomes**:

1. **Discharge Home / Care Facility** (60%)  
2. **Early Supported Discharge (ESD)** (40%) for further recovery at home.
